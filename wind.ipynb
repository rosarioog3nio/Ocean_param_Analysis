```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
wind_sul_east = pd.read_fwf('eastward_sul', header = None, sep='\s*,\s*', names = ['Tempo', 'non', '3', '33', 'Velocidade U'])
wind_sul_north = pd.read_fwf('northward_sul', header = None, sep='\s*,\s*', names = ['Tempo', 'non', '3', '33', 'Velocidade V'])
```

```python
wind_sul_north.head(3)
```

```python
wind_sul_east = pd.DataFrame(wind_sul_east)
wind_sul_north = pd.DataFrame(wind_sul_north)
type(wind_sul_north)
```

```python
wind_sul_east['Tempo'] = wind_sul_east['Tempo'].astype('datetime64')
wind_sul_north['Tempo'] = wind_sul_north['Tempo'].astype('datetime64')
wind_sul_east = wind_sul_east.set_index('Tempo')
wind_sul_north = wind_sul_north.set_index('Tempo')
```

```python
wind_sul_east = wind_sul_east.loc['2011-01-01': '2018-12-01']
wind_sul_north = wind_sul_north.loc['2011-01-01': '2018-12-01']
```

```python
wind_sul_east= wind_sul_east['Velocidade U']
dates_sul = wind_sul_east.index

wind_sul_north= wind_sul_north['Velocidade V']
dates_north = wind_sul_north.index
```

```python
ax3u = plt.plot(dates_sul, wind_sul_east, 'b', label='U Wind')
ax3v = plt.plot(wind_sul_north, 'r', label='V Wind')
plt.title('')
plt.ylabel('Northward Wind')
plt.xlabel('Eastward Wind')
plt.legend('')
plt.show()
```

```python
print('max = ',wind_sul_east.max(), wind_sul_east.idxmax())
print('min = ',wind_sul_east.min(), wind_sul_east.idxmin())

print('max = ',wind_sul_north.max(), wind_sul_north.idxmax())
print('min ',wind_sul_north.min(), wind_sul_north.idxmin())
```

```python
wind_norte_east = pd.read_fwf('eastward_norte.txt', header = None, sep='\s*,\s*', names = ['Tempo', 'non', 'Velocidade U'])
wind_norte_north = pd.read_fwf('northward_norte.txt', header = None, sep='\s*,\s*',names = ['Tempo', 'non', 'Velocidade V'])


wind_norte_east['Tempo'] = wind_norte_east['Tempo'].astype('datetime64')
wind_norte_north['Tempo'] = wind_norte_north['Tempo'].astype('datetime64')
wind_norte_east = wind_norte_east.set_index('Tempo')
wind_norte_north = wind_norte_north.set_index('Tempo')

#wind_norte_east = wind_norte_east.loc['2011-01-01': '2018-12-01']
#wind_norte_north = wind_norte_north.loc['2011-01-01': '2018-12-01']

wind_norte_east = pd.DataFrame(wind_norte_east)
wind_norte_north = pd.DataFrame(wind_norte_north)


wind_norte_east= wind_norte_east['Velocidade U']
dates_norte = wind_norte_east.index

wind_norte_north= wind_norte_north['Velocidade V']
dates_norte = wind_norte_north.index

axu1 = plt.plot(dates_norte, wind_norte_east, 'b', label='U Wind')
axv1 = plt.plot(wind_norte_north, 'r', label='V Wind')
plt.title('')
plt.ylabel('Northward Wind')
plt.xlabel('Eastward Wind')
plt.legend('')
plt.show()
```

```python
print('max = ',wind_norte_east.max(), wind_norte_east.idxmax())
print('min = ',wind_norte_east.min(), wind_norte_east.idxmin())

print('max = ',wind_norte_north.max(), wind_norte_north.idxmax())
print('min ',wind_norte_north.min(), wind_norte_north.idxmin())
```

```python
fig, (ax1,ax2) = plt.subplots(2, sharex=True, sharey=True, figsize=(20, 10))

ax1.set_axisbelow(True)
# Turn on the minor TICKS, which are required for the minor GRID
ax1.minorticks_on()


ax1.grid(which='major', linestyle='-', linewidth='0.5')
ax2.grid(which='major', linestyle='-', linewidth='0.5')


# Customize the minor grid
ax1.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
ax2.grid(which='minor', linestyle=':', linewidth='0.5', color='black')


fig.suptitle('Variacao do vento nas componetes U e V')
ax1.plot(dates_norte, wind_norte_east, 'b', label = 'U wind')
ax1.plot(wind_norte_north, 'r', label = 'V wind')
ax2.plot(dates_sul, wind_sul_east, 'b', label = 'U wind')
ax2.plot(wind_sul_north, 'r', label = 'V Wind')

ax1.legend(['Velocidade em U', 'Velocidade em V'])
ax2.legend(['Velocidade em U', 'Velocidade em V'])


ax2.set_xlabel('Tempo (Anos)')
ax2.set_ylabel('Velocidade (m/s)')
ax1.set_title('Zona Norte')
ax2.set_title('Zona Sul')
```
